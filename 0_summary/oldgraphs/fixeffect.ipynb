```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(-10, 10, 50)
m = np.linspace(-10, 0, 50)
line = x
line2 = -m

theta = np.linspace(0, 2*np.pi, 100)
r = np.sqrt(200)
r2 = np.sqrt(50)
r3 = np.sqrt(20)

x1 = r*np.cos(theta)
x2 = r*np.sin(theta)

x3 = r2*np.cos(theta) +10
x4 = r2*np.sin(theta) -10

x5 = r3*np.cos(theta) -10
x6 = r3*np.sin(theta) -10

x7 = r3*np.cos(theta) +10
x8 = r3*np.sin(theta) +10

plt.plot(x, line, color='blue', label='initial coordinates')
plt.plot(x1, x2, color='black', label='low-energy path')
plt.plot(x3, x4, color='red', label='oxygen')
plt.plot(x5, x6, color='orange', label='sodium')
plt.plot(x7, x8, color='orange')

plt.axvline(x=-10, ymin=(0.26), ymax=(0.82), color='black')
plt.axvline(x=-5, ymin=(0.40), ymax=(0.92), color='black')
plt.axvline(x=0, ymin=(0.54), ymax=(0.94), color='black')
plt.axvline(x=5, ymin=(0.68), ymax=(0.92), color='black')
plt.legend()
plt.axis('equal')
plt.savefig('/home/ben/Documents/gulp/fixeffect', bbox_inches='tight', dpi=500)
```

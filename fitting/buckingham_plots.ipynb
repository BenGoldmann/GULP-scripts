```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 3, 50)
initial = 1226.84 * np.exp((-x)/0.307)
khandy = 588.38 * np.exp((-x)/0.388)

plt.plot(x, initial, color='blue')
plt.plot(x, khandy, color='red')
plt.xlabel('separation, $\AA$')
plt.ylabel('energy, eV')
plt.title('Na-O Buckingham optimisation')
plt.legend(labels=['initial, 1226.84exp(-r/0.307)', 'new, 588.38exp(-r/0.388)'])
plt.savefig('/home/ben/Documents/gulp_calcs/0_summary/nao_fitting.jpg', bbox_inches='tight', dpi=1080)
```
